**Cleaning the Dataset**

Here we get rid of duplicates and null values. We also check for outliers and improperly handled data points and features.

```python
df.isna()
```

```python
nullV=df.isnull().sum()
print (nullV)
```

```python
#replacing the NaN values with a string since the column they are in 'Championship Years' has string values
#cannot simply use fillna
nan_replacements = {'Championship Years': 0.0}
df_clean = df.fillna(nan_replacements)
```

```python
cleaned_nullV=df_clean.isnull().sum()
print (cleaned_nullV)
```

```python
#checking for duplicates datapoints
duplicates=df.duplicated()
print(f"Number of duplicate rows: {duplicates.sum()}")
```

```python
#cant drop the rows with NA values here since they are all important drivers in the history of F1
#creating copy of dataframe so that original data is not tampered with
df1=df.copy()
```

```python
#outlier calculation using z score
z = np.abs(scipy.stats.zscore(df["Race_Starts"]))
outliers_count = np.sum(z > 3)

print(f"Number of outliers: {outliers_count}")
```

```python
#conversion of boolean value containing columns into numerical datapoints that infer the same value
df_clean['Champion'] = df_clean['Champion'].astype(int)
df_clean['Active'] = df_clean['Active'].astype(int)
df_clean
```

```python
#categorical data with string values that hold no significace if label/ordinal encoded
#that is if changed to numerical variables, have been dropped here to make a correlation matrix
#a correlation matrix measures strength and direction of relationships between pairs of variables in a dataset
corr = df_clean.drop(['Driver','Nationality','Seasons', 'Championship Years'], axis=1).corr()
plt.figure(figsize=(12, 10))

sns.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)],
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

```python
#unique values in the df_clean under Nationality since it is one of the most varied columns in this df
df_clean['Nationality'].unique()
```

```python
country_counts = df_clean['Nationality'].value_counts()
country_counts
```

```python
#to visually represent the countries better ive used a Choropleth map (via Plotly)
#it is best used to show values of variables that are patterned in proportion
#for geographical distribution of data
country_replacements = {
    'East Germany': 'Germany',
    'West Germany': 'Germany',
    'Rhodesia': 'Zimbabwe',
    'Rhodesia and Nyasaland': 'Zimbabwe',
    'RAF': 'United Kingdom',
    'Monaco Netherlands': 'Monaco',
    'Belgium France': 'Belgium'
}

country_counts.index = pd.Series(country_counts.index).replace(country_replacements)

#Standardising the names of the countries in order for choropeth to be able to
#recognise the country while adding the gradient
country_counts.index = country_counts.index.str.split(',').str[0]
country_counts.index = country_counts.index.str.strip().str.title()

# Group by the index to combine duplicates
country_counts = country_counts.groupby(country_counts.index).sum()

fig = px.choropleth(country_counts,
                    locations=country_counts.index,
                    locationmode='country names',
                    color=country_counts.values,
                    title="Where are the F1 Drivers From",
                    labels={'color':'Number of Drivers'})

fig.show()
```

```python
#now that we know the nationality split I have made the original df be consistenf with the new names
df_clean['Nationality'] = df_clean['Nationality'].replace(country_replacements).apply(lambda x: x.split(',')[0].strip())

#now we group by Nationality itself to analyze key statistics that give us insight champions belonging to each country
country = df_clean.groupby('Nationality')[['Championships','Race_Entries','Race_Starts','Pole_Positions','Race_Wins','Podiums']].sum()

fig = go.Figure()

fig.add_trace(go.Bar(y=country.index, x=country['Podiums'], name='Podiums',orientation='h'))
fig.add_trace(go.Bar(y=country.index, x=country['Championships'], name='Championships',orientation='h'))
fig.update_layout(barmode='stack', xaxis_title='Podiums and Championships', yaxis_title='Country', title='Podiums and Championships by Country')

fig.show()
```

```python
#defining a bar plot function so as to use it later to check the dependence of two variables on each other
# since Passing `palette` without assigning `hue` is deprecated and will be removed in v0.14.0. Assign the `x` variable to `hue` and set `legend=False` for the same effect.
# # instead of
# def barplot(col,col1,datas,color):
#     sns.barplot(x=col, y=col1 ,data=datas, palette=color, hue=None)
#     plt.show()
#ive used the following code to check if hue is necessary
def barplot(col, col1, data, palette):
    sns.barplot(x=col, y=col1, data=data, hue=col, palette=palette, legend=False)
    plt.legend([], [])
    plt.show()
```

```python
barplot('Championships','Points',df_clean,'Blues_d')
```

```python
barplot('Championships','Podiums',df_clean, 'Blues_d')
```

```python
#while the relation of points and podiums with championships is quite obvious, Pole_Positions vs Championships gives us a key insight into how well a driver is able to
#capitalise on their golden opportunity of starting first in a race
barplot('Championships','Pole_Positions',df_clean, 'plasma')
```

```python
#Boxplots are effective at highlighting outliers in the data.
# here when a driver has a high number of race starts but few championships, it could indicate consistency vs. excellence in winning
plt.figure(figsize = (10, 6))
ax = sns.boxplot(x='Championships', y='Race_Starts', data=df_clean)
plt.setp(ax.artists, alpha=.5, linewidth=2, edgecolor="k")
plt.xticks(rotation=45)
```
